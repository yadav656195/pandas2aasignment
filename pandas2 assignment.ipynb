```python
Q1. List any five functions of the pandas library with execution.

head() Function:

Displays the first n rows of a DataFrame.

import pandas as pd

# Sample DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

# Using head() to display the first 2 rows
print(df.head(2))
tail() Function:

Displays the last n rows of a DataFrame.

# Using tail() to display the last 1 row
print(df.tail(1))
describe() Function:

Generates descriptive statistics of a DataFrame.

# Using describe() to get summary statistics
print(df.describe())
info() Function:

Provides a concise summary of a DataFrame, including the data types and non-null values.

# Using info() to display DataFrame information
print(df.info())
value_counts() Function:

Counts the occurrences of unique values in a specified column.

# Sample DataFrame with a 'Category' column
df_category = pd.DataFrame({'Category': ['A', 'B', 'A', 'C', 'B']})

# Using value_counts() to count unique values in 'Category'
print(df_category['Category'].value_counts())
```

```python
Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.


import pandas as pd

def reindex_df(df):
    new_index = range(1, 2 * len(df) + 1, 2)
    df_reindexed = df.set_index(pd.Index(new_index))
    return df_reindexed

# Sample DataFrame
df = pd.DataFrame({'A': [10, 20, 30], 'B': [40, 50, 60], 'C': [70, 80, 90]})

# Call the function
df_reindexed = reindex_df(df)

# Display the re-indexed DataFrame
print(df_reindexed)
```

```python
Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the
sum of the first three values in the 'Values' column. The function should print the sum to the console.


import pandas as pd

def calculate_sum_of_first_three(df):
    values_sum = df['Values'].head(3).sum()
    print("Sum of the first three values:", values_sum)

# Sample DataFrame
df_values = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})

# Call the function
calculate_sum_of_first_three(df_values)
```

```python
Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.


import pandas as pd

def add_word_count_column(df):
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))
    return df

# Sample DataFrame
df_text = pd.DataFrame({'Text': ['Hello world', 'Python is awesome', 'Data Science']})

# Call the function
df_with_word_count = add_word_count_column(df_text)

# Display the DataFrame with the new 'Word_Count' column
print(df_with_word_count)
```

```python
Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.


import pandas as pd

def extract_username(df):
    df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])
    return df

# Sample DataFrame
df_emails = pd.DataFrame({'Email': ['john.doe@example.com', 'alice.smith@example.com']})

# Call the function
df_with_username = extract_username(df_emails)

# Display the DataFrame with the new 'Username' column
print(df_with_username)
Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows.

python
Copy code
import pandas as pd

def select_rows_condition(df):
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_rows

# Sample DataFrame
df_values = pd.DataFrame({'A': [3, 8, 6, 2, 9], 'B': [5, 2, 9, 3, 1], 'C': [1, 7, 4, 5, 2]})

# Call the function
selected_rows_df = select_rows_condition(df_values)

# Display the new DataFrame with selected rows
print(selected_rows_df)





```
